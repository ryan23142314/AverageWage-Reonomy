Question 1 : Find the Average Yearly Wages of H1-B Visa Aplicants in the San Francisco Proper and Manhattan

```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Pull in Data from CSV
data = pd.read_csv(r"C:\Users\ryana\Downloads\data_source (1).csv")
#Keep Necessary Columns
data = data[['visa_class','worksite_postal_code','worksite_state','total_workers','prevailing_wage','pw_unit_of_pay','wage_rate_of_pay_from']]
data.head()

```

```python
#Get only rows with H-1B Visa
data=data.loc[data['visa_class'] == "H-1B"]
#Since Manhattan has all zip codes that start with 100 and San Frsncisco has all zip codes that start with 941,
#We only need the first 3 characters of the code
data['worksite_postal_code'] = data['worksite_postal_code'].str.extract(r"(\d{3})")
data.head()
```

```python
#add yearly salary column
pay_type = [
    (data['pw_unit_of_pay'] == 'Month'),
    (data['pw_unit_of_pay'] == 'Year'),
    (data['pw_unit_of_pay'] == 'Hour'),
    (data['pw_unit_of_pay'] == 'Week'),
    (data['pw_unit_of_pay'] == 'Bi-Weekly')
    ]

# create a list of the values we want to assign for each condition
values = [data['prevailing_wage']*12, data['prevailing_wage']*1, data['prevailing_wage']*40*52, data['prevailing_wage']*52, data['prevailing_wage']*26]

# create a new column and use np.select to assign values to it using our lists as arguments
data['YearlyWage'] = np.select(pay_type, values)

# display updated DataFrame
data.head()
```

```python
#create seperate models for San Francisco and Manhattan
NewYork = data[data['worksite_postal_code'] == '100']
NewYork = NewYork[NewYork['worksite_state'] == 'NY']
SanFran = data[data['worksite_postal_code'] == '941']
SanFran = SanFran[SanFran['worksite_state'] == 'CA']
```

```python
#Find Average Yearly Salaries
AverageNY = (NewYork['YearlyWage']*NewYork['total_workers']).sum()/NewYork['total_workers'].sum()
print("The Average wage for H-1B applicant in Manhattan is ", AverageNY)
AverageSF = (SanFran['YearlyWage']*SanFran['total_workers']).sum()/SanFran['total_workers'].sum()
print("The Average wage for H-1B applicant in San Francisco is ", AverageSF)
```

Question 2 : What causes the difference in wages

Theory 1 : New Yorkers pay for Taxes and Fees before their prevailing wage


```python
#Find Average percent of prevailng wage out of total wage

AverageNYRate = NewYork['prevailing_wage'].sum()/NewYork['wage_rate_of_pay_from'].sum()
AverageSFRate = SanFran['prevailing_wage'].sum()/SanFran['wage_rate_of_pay_from'].sum()
print("Average amount of wage that employees recieve in Manhattan ", AverageNYRate)
print("Average amount of wage that employees recieve in San Francisco ", AverageSFRate)
```

This shows that in San Francisco employees take home more of their total wage
